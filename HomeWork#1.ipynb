### 1. a="Hello Python" 문자열을 Slicing과 연결 연산자 (+)를 사용하여 "Python Hello"로 변경하시오.
- a 변수 자체에 "Python Hello" 문자열이 저장되어야 함

```python
a="Hello Python"
tmp = a
a = tmp[6:] + " "+tmp[:6]
print(a)
```

### 8. 사용자로 부터 정수를 입력받아서 1부터 그 사이에 존재하는 소수 (Prime number)를 출력하는 파이썬 프로그램을 작성하시오.

```python
def get_prime_nums1(n):
    prime_nums = []

    for i in range(2, n + 1):
        flag = 1
        for j in range(2, i // 2 + 1):
            if i % j == 0:
                flag = 0
                break

        if flag == 1:
            prime_nums.append(i)

    return prime_nums
    
    
if __name__ == '__main__':

	print(get_prime_nums1(10)) # [2, 3, 5, 7]
```
